```python
df=data
```

```python
# converting the objects into int
df["Education_Level"]=df["Education_Level"].map({"High School":0,"Bachelor's":1,"Master's":2,"Doctorate":3})
df["Occupation"]=df["Occupation"].map({"Healthcare":0,"Education":1,"Technology":2,"Finance":3,"Others":4})
df["Location"]=df["Location"].map({"Urban":0,"Suburban":1,"Rural":2})
df["Marital_Status"]=df["Marital_Status"].map({"Single":0,"Married":1,"Divorced":2})
df["Employment_Status"]=df["Employment_Status"].map({"Full-time":0,"Part-time":1,"Self-employed":2})
df["Homeownership_Status"]=df["Homeownership_Status"].map({"Own":0,"Rent":1})
df["Type_of_Housing"]=df["Type_of_Housing"].map({"Apartment":0,"Single-family home":1,"Townhouse":2})
df["Gender"]=df["Gender"].map({"Male":0,"Female":1})
df["Primary_Mode_of_Transportatio"]=df["Primary_Mode_of_Transportation"].map({"Car":0,"Public transit":1,"Biking":2,"Walking":3})
```

```python
#Normalizing the data
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler 
scaler= MinMaxScaler()
```

```python
df['Age']=scaler.fit_transform(df[['Age']])
df['Work_Experience']=scaler.fit_transform(df[['Work_Experience']])
```

**The selected models are Linear Regression,Decision Tree and XG boost**

```python
#for processesed data required models
features=['Age','Education_Level','Occupation','Number_of_Dependents','Location','Work_Experience','Marital_Status','Employment_Status','Household_Size','Homeownership_Status','Type_of_Housing','Gender','Primary_Mode_of_Transportatio']
x=df[features].values
y=df['Income'].values
```

```python
#for models doesn't required preprocessing
X=data[features].values
Y=data['Income'].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.30)
```

```python
import xgboost as xgb
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
```

```python
model_xgb=xgb.XGBRegressor()
model_lr= LinearRegression()
model_dt = DecisionTreeRegressor()
```

```python
model_xgb.fit(X_train,Y_train)
model_lr.fit(x_train,y_train)
model_dt.fit(X_train,Y_train)
```

```python
import numpy as np
y_pred_lr = model_lr.predict(x_test)
y_pred_xgb = model_lr.predict(X_test)
y_pred_dt= model_lr.predict(X_test)
```

**Evaluate MSE, RMSE, R2 Score**

```python
#for Liniar regression
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred_lr)
print("MSE:", mse)

rmse = np.sqrt(mse)
print("RMSE:", rmse)

r2 = r2_score(y_test, y_pred_lr)
print("R2:", r2)
```

```python
#for decision tree
mse = mean_squared_error(Y_test, y_pred_dt)
print("MSE:", mse)

rmse = np.sqrt(mse)
print("RMSE:", rmse)

r2 = r2_score(Y_test, y_pred_dt)
print("R2:", r2)
```

```python
#for XGboost
mse = mean_squared_error(Y_test, y_pred_xgb)
print("MSE:", mse)

rmse = np.sqrt(mse)
print("RMSE:", rmse)

r2 = r2_score(Y_test, y_pred_xgb)
print("R2:", r2)
```

**Models with the lowest RMSE and MSE are preferred, as they indicate that the model's predictions are closer to the actual values.A model with a higher R² score is better at explaining the variance in the data.**

 decision tree Model and xg boost Model  are the best models among the three based on the provided evaluation metrics,selecting xg boost model to save
