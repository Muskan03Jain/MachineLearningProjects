**Removing unrelevent columns**

```python
data=data.drop(['Surname'],axis=1)
 
```

```python
data.drop(['CustomerId'],axis=1,inplace=True) 
```

**Handling nan values**

```python
data.isna().sum()
```

```python
data=data.dropna(axis=0)
data=data.drop("RowNumber",axis=1)
```

**Normalizing the data**

```python
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler 
scaler= MinMaxScaler()
```

```python
data.describe()
```

```python
#|for the model required encoding and normalized data
df=data
```

```python
df['CreditScore']=scaler.fit_transform(df[['CreditScore']])
df['Age']=scaler.fit_transform(df[['Age']])
df['Balance']=scaler.fit_transform(df[['Balance']])
df['EstimatedSalary']=scaler.fit_transform(df[['EstimatedSalary']])
```

```python
df.describe()
```

**Encoding the data**

```python
df.head()
```

```python
df['Geography']=df['Geography'].map({'France':0,'Spain':1,'Germany':2})
df['Gender']=df['Gender'].map({'Male':0,'Female':1})
```

```python
df.head()
```

```python
df['Exited'].value_counts()
```

**comparing the different models which is KNN,Bagging and boosting algorithms**

```python
features=['CreditScore','Geography','Gender','Age','Tenure','Balance','NumOfProducts','HasCrCard','IsActiveMember','EstimatedSalary']
x=df[features].values
y=df['Exited'].values
```

```python
#data without preocessing
features=['CreditScore','Geography','Gender','Age','Tenure','Balance','NumOfProducts','HasCrCard','IsActiveMember','EstimatedSalary']
X=data[features].values
Y=data['Exited'].values
```

```python
from sklearn.model_selection import train_test_split
#for processesed data required model
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
#for the model which doesnot required preprocessing
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
model_KNN = KNeighborsClassifier()
model_RC =  RandomForestClassifier()
model_GB = GradientBoostingClassifier()
```

```python
model_KNN.fit(x_train,y_train)
model_RC.fit(X_train,Y_train)
model_GB.fit(X_train,Y_train)
```

```python
y_pred_knn = model_KNN.predict(x_test)
y_pred_rc = model_RC.predict(X_test)
y_pred_gb = model_GB.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score 
```

```python
print("KNN accuracy: ",accuracy_score(y_test , y_pred_knn)*100)
print("Random Forest accuracy: ",accuracy_score(Y_test , y_pred_rc)*100)
print("Boosting accuracy: ",accuracy_score(Y_test , y_pred_gb)*100)
```

```python
from sklearn.metrics import recall_score,precision_score,f1_score
```

```python
#Recall
print("KNN Recall : ",recall_score(y_test,y_pred_knn)*100)
print("Random Forest Recall : ",recall_score(Y_test,y_pred_rc)*100)
print("Boosting Recall : ",recall_score(Y_test,y_pred_gb)*100)
```

```python
#precision
print("KNN precision : ",precision_score(y_test,y_pred_knn)*100)
print("Random Forest precision : ",precision_score(Y_test,y_pred_rc)*100)
print("Boosting precision : ",precision_score(Y_test,y_pred_gb)*100)
```

```python
#F1 Score
print("KNN f1 : ",f1_score(y_test,y_pred_knn)*100)
print("Random Forest f1 : ",f1_score(Y_test,y_pred_rc)*100)
print("Boosting f1 : ",f1_score(Y_test,y_pred_gb)*100)
```

**In this system it is found more dangerous if the customer is not exited but prediction is exited so in this case we use precision score to deside the best model among all which is Gradient Boosting Algorithm**
