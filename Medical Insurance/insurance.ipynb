```python
import pandas as pd
import numpy as np
```

```python
#loading dataset
insurance_data = pd.read_csv(r"C:\Users\Lenovo\Muskan\Ml\github assignment\Machine-Learning-Projects\Linear_Regression\Medical_Insurance\insurance.csv")
```

```python
insurance_data.head()
```

```python
insurance_data.isna().sum()
```

```python
insurance_data['sex'] = insurance_data['sex'].map({'male': 0, 'female': 1})
```

```python
insurance_data.head()
```

```python
insurance_data['smoker'] = insurance_data['smoker'].map({'yes': 0, 'no': 1})
```

```python
insurance_data.head()
```

```python
insurance_data['region'] = insurance_data['region'].map({'southwest': 0, 'southeast': 1,'northwest':2,'northeast':3})
```

```python
insurance_data.head()
```

```python
insurance_data.info()
```

```python
features=['age','sex','bmi','children','smoker','region']
insurance_data[features].values
```

```python
X= insurance_data[features].values
Y=insurance_data['charges'].values
```

```python
type(X)
type(Y)
```

```python
from sklearn.model_selection import train_test_split
# Split data 70%-30% into training set and test set
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.30)
```

```python
Y_test
```

```python
# Train the model
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression().fit(X_train, Y_train)

print (model)
```

```python
predictions = model.predict(X_test)
```

```python
 #predicted
print('Predicted labels: ', np.round(predictions, 0)[:10])
print('Actual labels   : ' ,Y_test[:10])
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(Y_test, predictions)
print("MSE:", mse)

rmse = np.sqrt(mse)
print("RMSE:", rmse)

r2 = r2_score(Y_test, predictions)
print("R2:", r2)
```
